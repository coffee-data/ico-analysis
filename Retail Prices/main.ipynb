```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('retail_prices_cleaned.csv')
df = df.rename(columns=({'Calendar years':'Region'}))
df.head()
```

```python
# Drop soluble coffee data for separate df.
dfc = df[df['Region'] != 'United Kingdom 1']

for year in dfc.loc[:,'1990':'2018']:
    dfc[year] = (dfc[year]*100)

dfc
```

```python
df1 = dfc.dropna()
df1
```

```python
# Find max and min in each row and set to list to check for total max.
lst_max = []
lst_min = []

for col in df1.loc[:,'1990':'2018']:
    s = df1[col]
    lst_max.append(max(s))
    lst_min.append(min(s))

y_lim_max = int(max(lst_max) + min(lst_max)*.2)
y_lim_min = int(min(lst_min) - min(lst_min)*.2)

print(y_lim_max)
```

```python
def clean(data):
    data = data.transpose()
    new_header = data.iloc[0]
    new_header
    data = data[1:]
    data.columns = new_header
    data = data.reset_index()
    data = data.rename(columns={"index":"Year"})
    return data

def ico_rp_plot_all(df, f1, f2, graph_name, l_lim, u_lim, s_year, e_year):
    
    # Define Colors
    # These are the "Tableau 20" colors as RGB.  
    tableau20 = [(31, 119, 180), (174, 199, 232), (255, 127, 14), (255, 187, 120),  
                 (44, 160, 44), (152, 223, 138), (214, 39, 40), (255, 152, 150),  
                 (148, 103, 189), (197, 176, 213), (140, 86, 75), (196, 156, 148),  
                 (227, 119, 194), (247, 182, 210), (127, 127, 127), (199, 199, 199),  
                 (188, 189, 34), (219, 219, 141), (23, 190, 207), (158, 218, 229)]  

    # Scale the RGB values to the [0, 1] range, which is the format matplotlib accepts.  
    for i in range(len(tableau20)):  
        r, g, b = tableau20[i]  
        tableau20[i] = (r / 255., g / 255., b / 255.)  
    
    # Convert years to int64
    df.Year = df.Year.astype('int64')
    df.Year
    
    # 1. Declare fig size
    plt.figure(figsize=(f1, f2)) 


    # 2. Remove all spines. Considered chartjunk
    ax = plt.subplot(111)  
    ax.spines["top"].set_visible(False)  
    ax.spines["bottom"].set_visible(False)  
    ax.spines["right"].set_visible(False)  
    ax.spines["left"].set_visible(False)  


    # 3. Set ticks to bottom and left only
    ax.get_xaxis().tick_bottom()  
    ax.get_yaxis().tick_left()  


    # 4. Set limits
    plt.ylim(l_lim, u_lim)
    plt.xlim(s_year, e_year)


    # 5. Plot ticks
    plt.yticks(range(100, (u_lim + 1), 100), fontsize=14)  
    plt.xticks(range(s_year, e_year, 1), rotation=45, fontsize=14)  


    # 6. Plot tick lines
    for y in range(100, u_lim+100, 100):  
        plt.plot(range(s_year, e_year), [y] * len(range(s_year, e_year)), "--", lw=0.5, color="black", alpha=0.3)  


    lst_regions = []
    years = df['Year']
    x_mean = years.mean()
    max_year = int(max(df.Year.values))
    min_year = int(min(df.Year.values))


    # Set list for regions and exclude year column.
    for col_name in df:
        if col_name != 'Year':
            lst_regions.append(col_name)


    # Iterate over the index, and for each region, plt the region's values against years.
    for i, j in enumerate(lst_regions):
        plt.plot(df.Year.values,
                 df[j].values,
                lw=2.5, color=tableau20[i])


        # Take the last value on the plot, adjust the coordinates, and place text
        y_pos = df[j].values[-1] - 0.5
        text = plt.text(2018.5, y_pos, j, fontsize=12, color=tableau20[i])


    # Custom title built to center on mean year
    plt.text(x_mean, u_lim + 100, f"{graph_name}"  
           f", ({s_year}-{e_year})", fontsize=17, ha="center")  


    # Custom footer text
    plt.text((s_year - 1), (l_lim - 200), "Data Source: http://www.ico.org/new_historical.asp"  
           "\nAuthor: Jesse Hartman (thecoffeepodcast.com/coffee-data / @coffee.data)"  
           "\nNote: Regions with data gaps were ommitted from this analysis.", fontsize=10)


    plt.ylabel('US Cents per LB', fontsize=14)


    # The magic of tightness
    plt.tight_layout()


    # Save figure
    plt.savefig(f"output_files/{(graph_name.lower().replace(' ', '-'))}-{s_year}-{e_year}.png", bbx_inches='tight')
    plt.show()
                
def ico_rp_plot_10yr(df, f1, f2, graph_name, l_lim, u_lim, s_year, e_year):
    # Function for 10 year with lower limit and upper limit
    
    # Define Colors
    # These are the "Tableau 20" colors as RGB.  
    tableau20 = [(31, 119, 180), (174, 199, 232), (255, 127, 14), (255, 187, 120),  
                 (44, 160, 44), (152, 223, 138), (214, 39, 40), (255, 152, 150),  
                 (148, 103, 189), (197, 176, 213), (140, 86, 75), (196, 156, 148),  
                 (227, 119, 194), (247, 182, 210), (127, 127, 127), (199, 199, 199),  
                 (188, 189, 34), (219, 219, 141), (23, 190, 207), (158, 218, 229)]  

    # Scale the RGB values to the [0, 1] range, which is the format matplotlib accepts.  
    for i in range(len(tableau20)):  
        r, g, b = tableau20[i]  
        tableau20[i] = (r / 255., g / 255., b / 255.)  
    
    # Convert years to int64
    df.Year = df.Year.astype('int64')
    df.Year
    
    # 1. Declare fig size
    plt.figure(figsize=(f1, f2)) 


    # 2. Remove all spines. Considered chartjunk
    ax = plt.subplot(111)  
    ax.spines["top"].set_visible(False)  
    ax.spines["bottom"].set_visible(False)  
    ax.spines["right"].set_visible(False)  
    ax.spines["left"].set_visible(False)  


    # 3. Set ticks to bottom and left only
    ax.get_xaxis().tick_bottom()  
    ax.get_yaxis().tick_left()  


    # 4. Set limits
    plt.ylim(l_lim, u_lim)
    plt.xlim(1998, 2019)


    # 5. Plot ticks
    plt.yticks(range(l_lim, u_lim, 50), fontsize=14)  
    plt.xticks(range(s_year, e_year, 1), rotation=45, fontsize=14)  


    # 6. Plot tick lines
    for y in range(l_lim, u_lim, 50):  
        plt.plot(range(s_year, e_year), [y] * len(range(s_year, e_year)), "--", lw=0.5, color="black", alpha=0.3)  


    lst_regions = []
    years = df['Year']
    x_mean = years.mean()
    max_year = int(max(df.Year.values))
    min_year = int(min(df.Year.values))


    # Set list for regions and exclude year column.
    for col_name in df:
        if col_name != 'Year':
            lst_regions.append(col_name)


    # Iterate over the index, and for each region, plt the region's values against years.
    for i, j in enumerate(lst_regions):
        plt.plot(df.Year.values,
                 df[j].values,
                lw=2.5, color=tableau20[i])


        # Take the last value on the plot, adjust the coordinates, and place text
        y_pos = df[j].values[-1] - 0.5
        if j == "Norway":  
            y_pos -= 10  
        elif j == "Cyprus":  
            y_pos += 10  
        elif j == "Denmark":  
            y_pos -= 10  
        text = plt.text(2018.5, y_pos, j, fontsize=12, color=tableau20[i])


    # Custom title built to center on mean year
    plt.text(x_mean, u_lim + 100, f"{graph_name}"  
           f", ({s_year}-{e_year})", fontsize=17, ha="center")  


    # Custom footer text
    plt.text((s_year - 1), - (l_lim - 100), "Data Source: http://www.ico.org/new_historical.asp"  
           "\nAuthor: Jesse Hartman (thecoffeepodcast.com/coffee-data / @coffee.data)"  
           "\nNote: Regions with data gaps were ommitted from this analysis.", fontsize=10)


    plt.ylabel('US Cents per LB', fontsize=14)


    # The magic of tightness
    plt.tight_layout()


    # Save figure
    plt.savefig(f"output_files/{(graph_name.lower().replace(' ', '-'))}-{s_year}-{e_year}.png", bbx_inches='tight')
    plt.show()
```
