```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Function to clean data
def clean(data):
    data = data.transpose()
    new_header = data.iloc[0]
    new_header
    data = data[1:]
    data.columns = new_header
    data = data.reset_index()
    data = data.rename(columns={"index":"Year"})
    return data
```

```python
df_cm = pd.read_csv("cleaned_data/colombians_milds.csv")
df_cm.head()
# Drop any row with missing values
df_cm = df_cm.dropna()
df_cm
```

```python
# Run clean function
df_cm = clean(df_cm)
df_cm.head()
```

```python
# Export table to html
df_cm.to_html('tbl_col_milds.html')
```

```python
# Define Colors
# These are the "Tableau 20" colors as RGB.  
tableau20 = [(31, 119, 180), (174, 199, 232), (255, 127, 14), (255, 187, 120),  
             (44, 160, 44), (152, 223, 138), (214, 39, 40), (255, 152, 150),  
             (148, 103, 189), (197, 176, 213), (140, 86, 75), (196, 156, 148),  
             (227, 119, 194), (247, 182, 210), (127, 127, 127), (199, 199, 199),  
             (188, 189, 34), (219, 219, 141), (23, 190, 207), (158, 218, 229)]  

# Scale the RGB values to the [0, 1] range, which is the format matplotlib accepts.  
for i in range(len(tableau20)):  
    r, g, b = tableau20[i]  
    tableau20[i] = (r / 255., g / 255., b / 255.)  
```

# Below is code for graphing the only complete set of Colombian Milds.

```python
# Set values for x and y and make sure they are integers 
# to reference axis placement for later
x = df_cm["Year"].astype('int32')
y = df_cm['Colombia'].astype('int32')
coffee_type = 'Colombian Milds'
coffee_region = 'Colombia'
lst_coffee_region = []
```

```python
def coffee_plt (df, x, y, coffee_type, coffee_region):

    # Y position at 2018 for graph purposes
    for elt in df[coffee_region]:
        y_pos = elt

    # Set length values
    x_len = len(x)

    # Set figure size
    plt.figure(figsize=(12, 10))  

    # Remove the plot frame lines. They are unnecessary chartjunk.  
    ax = plt.subplot(111)  
    ax.spines["top"].set_visible(False)  
    ax.spines["bottom"].set_visible(True)  
    ax.spines["right"].set_visible(False)  
    ax.spines["left"].set_visible(False)  

    # Limit the range of the plot to only where the data is.  
    plt.ylim(0, max(y)+30)
    plt.xlim(1990, 2018)  

    # Plot tick lines
    for j in range(0, max(y)+40, 10):  
        plt.plot(range(1990, 2019), [j] * len(range(1990, 2019)), "--", lw=0.5, color="black", alpha=0.3)  

    # Remove the tick marks; they are unnecessary with the tick lines we just plotted.  
    plt.tick_params(axis="both", which="both", bottom="off", top="off",  
                    labelbottom="on", left="off", right="off", labelleft="on")  

    # Plot data
    plt.plot(x, y, lw=2.5, color=tableau20[0])

    plt.yticks(range(0, max(y)+40, 10), [str(x) + "" for x in range(0, max(y)+40, 10)], fontsize=14)  
    plt.xticks(range(1990,2019,1), rotation=45, fontsize=14) 

    # Ensure that the axis ticks only show up on the bottom and left of the plot.  
    # Ticks on the right and top of the plot are generally unnecessary chartjunk.  
    ax.get_xaxis().tick_bottom()  
    ax.get_yaxis().tick_left()  

    # Set mean for x axis to get center of graph for txt.
    x_mean = x.mean()

    plt.text(x_mean, max(y) + 40, f"Prices to Growers - Annual Averages"  
           f", {coffee_type} (1990-2018)", fontsize=17, ha="center")  

    plt.text(1989, -75, "Data Source: http://www.ico.org/new_historical.asp"  
           "\nAuthor: Jesse Hartman (thecoffeepodcast.com/coffee-data / @coffee.data)"  
           "\nNote: Historical data before 1990 was behind a paywall.", fontsize=10)

    plt.text(2018.5, y_pos, coffee_region, fontsize=14, color=tableau20[0]) 

    plt.ylabel('US Cents per LB', fontsize=14)

    plt.show()
```

```python
# Test coffee_plt with Colombian Milds
coffee_plt(df_cm, x, y, coffee_type, coffee_region)
```

```python
# Function for plotting all regions
def ico_fg_plot_all(df, coffee_type):
    
    # Convert years to int64
    df.Year = df.Year.astype('int64')
    df.Year
    
    # 1. Declare fig size
    plt.figure(figsize=(12, 9)) 

    # 2. Remove all spines. Considered chartjunk
    ax = plt.subplot(111)  
    ax.spines["top"].set_visible(False)  
    ax.spines["bottom"].set_visible(False)  
    ax.spines["right"].set_visible(False)  
    ax.spines["left"].set_visible(False)  

    # 3. Set ticks to bottom and left only
    ax.get_xaxis().tick_bottom()  
    ax.get_yaxis().tick_left()  

    # 4. Set limits
    plt.ylim(0, 250)
    plt.xlim(1990, 2019)

    # 5. Plot ticks
    plt.yticks(range(0, 251, 10), fontsize=14)  
    plt.xticks(range(1990, 2019, 1), rotation=45, fontsize=14)  

    # 6. Plot tick lines
    for y in range(0, 251, 10):  
        plt.plot(range(1990, 2019), [y] * len(range(1990, 2019)), "--", lw=0.5, color="black", alpha=0.3)  

    lst_regions = []
    years = df['Year']
    x_mean = years.mean()
    max_year = int(max(df.Year.values))
    min_year = int(min(df.Year.values))

    # Set list for regions and exclude year column.
    for col_name in df:
        if col_name != 'Year':
            lst_regions.append(col_name)

    # Iterate over the index, and for each region, plt the region's values against years.
    for i, j in enumerate(lst_regions):
        plt.plot(df.Year.values,
                 df[j].values,
                lw=2.5, color=tableau20[i])

        # Take the last value on the plot, adjust the coordinates, and place text
        y_pos = df[j].values[-1] - 0.5
        text = plt.text(2018.5, y_pos, j, fontsize=12, color=tableau20[i])

    # Custom title built to center on mean year
    plt.text(x_mean, 255, f"Prices to Growers - Annual Averages"  
           f", {coffee_type} (1990-2018)", fontsize=17, ha="center")  

    # Custom footer text
    plt.text(1989, -50, "Data Source: http://www.ico.org/new_historical.asp"  
           "\nAuthor: Jesse Hartman (thecoffeepodcast.com/coffee-data / @coffee.data)"  
           "\nNote: Regions with data gaps were ommitted from this analysis.", fontsize=10)

    plt.ylabel('US Cents per LB', fontsize=14)

    # The magic of tightness
    plt.tight_layout()

    # Save figure
    plt.savefig(f"output_files/{(coffee_type.lower().replace(' ', '-'))}-1990-2018.png", bbx_inches='tight')
    plt.show()
```
